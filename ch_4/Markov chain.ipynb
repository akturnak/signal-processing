```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
n = 1000
```

```python
sigma = 1
```

```python
f_noise = np.random.normal(0, sigma, n)
```

```python
f_signal = np.zeros(n)
```

```python
for i in range(1, n):
    f_signal[i] = f_signal[i-1] + f_noise[i]
```

```python
plt.plot(f_noise)
```

```python
plt.plot(f_signal)
```

```python
plt.grid(True)
```

```python
plt.show()
```
