```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
n = 1000
```

```python
sigma = 5
```

```python
r = 0.99
```

```python
en = np.sqrt((1-r*r)*sigma*sigma)
```

```python
f_signal = np.zeros(n)
```

```python
f_signal[0] = np.random.normal(0, sigma)
```

```python
for i in range(1, n):
    f_signal[i] = r*f_signal[i-1] + np.random.normal(0, en)
```

```python
plt.plot(f_signal)
plt.grid(True)
plt.show()
```
